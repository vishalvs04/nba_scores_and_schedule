```python
def get_scoreboard():
    scoreboard=get_links()['currentScoreboard']
    games=get(base_url+scoreboard).json()['games']
    for game in games:
        home_team=game['hTeam']
        away_team=game['vTeam']
        clock=game['clock']
        period=game['period']
        print(f"{home_team['triCode']} vs {away_team['triCode']}")
        print(f"{home_team['score']} - {away_team['score']}")
        print(f"{clock} and {period['current']}")
        print('------------------------------------------------------')
```

```python
def get_links():
    data=get(base_url+all_json).json()
    links=data['links']
    return links
```

```python
def get_scoreboards():
    scoreboard=get_links()['currentScoreboard']
    games=get(base_url+scoreboard).json()['games']
    for game in games:
        print(game)
```

```python
def get_teams():
    teams_url=get_links()['teams']
    teams=get(base_url+teams_url).json()
    new_teams_dict={}
    cities=['vegas', 'standard', 'utah', 'africa', 'sacramento']
    for city in cities:
        for team in teams['league'][city]:
            new_teams_dict[team['teamId']]=team['fullName']
        return new_teams_dict
```

```python
def get_league_schedule():
    schedule_url=get_links()['leagueSchedule']
    schedule_data=get(base_url+schedule_url).json()['league']['standard']
    teams=get_teams()
    for game in schedule_data:
        print(f"Game: {game['gameId']}, Time: {game['startTimeEastern']} ")
        print(f"{teams[game['hTeam']['teamId']]}    VS    {teams[game['vTeam']['teamId']]}")
        print(f"\n{teams[game['hTeam']['teamId']]} Stats: WINS:{game['hTeam']['win']} and LOSS:{game['hTeam']['loss']}")
        print(f"{teams[game['vTeam']['teamId']]} Stats: WINS:{game['vTeam']['win']} and LOSS:{game['vTeam']['loss']}")
        if game['hTeam']['score']>game['vTeam']['score']:
            print(f"\nMatch Score:\n{teams[game['hTeam']['teamId']]}:{game['hTeam']['score']}\n{teams[game['vTeam']['teamId']]}:{game['vTeam']['score']}")
            print("Winner:",teams[game['hTeam']['teamId']])
        elif game['hTeam']['score']<game['vTeam']['score']:
            print(f"\nMatch Score:\n{teams[game['hTeam']['teamId']]}:{game['hTeam']['score']}\n{teams[game['vTeam']['teamId']]}:{game['vTeam']['score']}")
            print("Winner:",teams[game['vTeam']['teamId']])
        elif game['hTeam']['score']=='' and game['vTeam']['score']=='':
            print(f"\nMatch Score:\n{teams[game['hTeam']['teamId']]}:Null Game hasn't begun\n{teams[game['vTeam']['teamId']]}:Null Game hasn't begun")
        print('_____________________________________________________________________________')
        
```
